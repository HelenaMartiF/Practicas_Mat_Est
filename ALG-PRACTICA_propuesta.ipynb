### 1.2 Implementación del modelo

Lo primero, implementa una función que devuelva los valores de $\widehat{w}$ dados X e y. Si se da el caso de que el dataset tiene más features que observaciones, queremos transponerlo antes de realizar las operaciones. Lo hacemos porque, como vimos en clase, los resultados no van a ser buenos si el número de observaciones es menor.

```python
import numpy as np

def pesos_ols(X, y):
    # Implementa la fórmula matricial
    
    # TODO 1 Comprueba que las dimensiones son adecuadas y corrige si no es así

    if X.shape[0] < X.shape[1]:
        X = X.T

    if y.ndim == 1:
        y = y.reshape(-1, 1)
    
    # TODO 2 Prepara la matriz X para que tenga la forma adecuada para aplicar la fórmula 

    if X.shape[0] != y.shape[0]:
        raise ValueError("El número de filas en X debe coincidir con el número de filas en y")
    

    # TODO 3 Calcula los pesos
    X_T_X = X.T @ X
    X_T_X_inv = np.linalg.inv(X_T_X)
    X_T_y = X.T @ y
    w = X_T_X_inv @ X_T_y

    return w.flatten()
```

## 2. Optimización con descenso del gradiente

Vamos a resolver el mismo problema usando el descenso del gradiente en lugar de la solución analítica. El descenso del gradiente es un método de optimización iterativo que usa el operador gradiente.

Como es iterativo, necesitamos decirle cuándo tiene que parar y darnos la solución. En este caso, le vamos a decir que pare y nos devuelva la solución después de un número de iteraciones que le pasaremos como parámetro. La ecuación iterativa es.

$$w^{t+1} = w^t - \eta \cdot \nabla f(w)$$

donde f es nuestra función objetivo, y w es un vector. Para la función objetivo, vamos a usar el error RSS, que querremos minimizar.

$$ RSS(w) = \frac{1}{2}\sum_{n=1}^{N}[y_n-f(x_n)]^2$$

y cuyo gradiente es:

$$\nabla RSS(w) = X^T(Xw^t-y)$$

TIP: Ten en cuenta que el RSS también lo puedes escribir como $\frac{1}{2}\sum_{n=1}^{N}[y_n-\hat{y_n}]^2$ donde $\hat{y_n} = X \hat{w}$ en cada paso.

### 2.0 Compara el método de mínimos cuadrados y el descenso del gradiente para obtener los coeficientes de una regresión lineal



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos proporcionados (reemplaza esto con tus datos reales)
df = pd.DataFrame({
    'feature_column': [1710, 1262, 1786, 1717, 2198, 1362, 1694, 2090, 1774, 1077],
    'target_column': [208500, 181500, 223500, 140000, 250000, 143000, 307000, 200000, 129900, 118000]
})

# Preparar los datos
X = df[['feature_column']].values
y = df['target_column'].values

# Agregar un término de intercepto (1's) a X
X_with_intercept = np.hstack([np.ones((X.shape[0], 1)), X])  # Agrega una columna de unos

# Función para calcular los pesos usando OLS
def pesos_ols(X, y):
    if X.shape[0] < X.shape[1]:
        X = X.T
    y = y.reshape(-1, 1) if y.ndim == 1 else y
    X_T_X = X.T @ X
    X_T_X_inv = np.linalg.inv(X_T_X)
    X_T_y = X.T @ y
    w = X_T_X_inv @ X_T_y
    return w.flatten()

# Calcular los pesos usando nuestra función
w_ols = pesos_ols(X_with_intercept, y)
intercepto_ols, pendiente_ols = w_ols

# Implementar el descenso del gradiente
def descenso_gradiente(X, y, eta=0.01, iteraciones=1000):
    m, n = X.shape
    w = np.zeros(n)
    for _ in range(iteraciones):
        gradient = X.T @ (X @ w - y) / m
        w -= eta * gradient
    return w

# Calcular los pesos usando descenso del gradiente
w_gradiente = descenso_gradiente(X_with_intercept, y, eta=0.00000001, iteraciones=10000)
intercepto_gradiente, pendiente_gradiente = w_gradiente

# Ajustar el modelo usando sklearn
lr = LinearRegression()
lr.fit(X, y)
intercepto_sklearn = lr.intercept_
pendiente_sklearn = lr.coef_[0]

# Imprimir resultados
print("Resultados de la función pesos_ols:")
print(f"Punto de corte (intercepto): {intercepto_ols}")
print(f"Pendiente: {pendiente_ols}")

print("\nResultados de descenso del gradiente:")
print(f"Punto de corte (intercepto): {intercepto_gradiente}")
print(f"Pendiente: {pendiente_gradiente}")

print("\nResultados de sklearn:")
print(f"Punto de corte (intercepto): {intercepto_sklearn}")
print(f"Pendiente: {pendiente_sklearn}")

# Graficar los resultados
plt.figure(figsize=(10, 6))

# Graficar los datos reales
plt.scatter(X, y, color='blue', label='Datos Reales')

# Línea de ajuste con nuestra función
plt.plot(X, intercepto_ols + pendiente_ols * X, color='red', label='Línea de Regresión (pesos_ols)')

# Línea de ajuste con descenso del gradiente
plt.plot(X, intercepto_gradiente + pendiente_gradiente * X, color='orange', linestyle='-.', label='Línea de Regresión (descenso_gradiente)')

# Línea de ajuste con sklearn
plt.plot(X, intercepto_sklearn + pendiente_sklearn * X, color='green', linestyle='--', label='Línea de Regresión (sklearn)')

plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Comparación de Regresiones Lineales')
plt.legend()
plt.grid(True)
plt.show()
```
